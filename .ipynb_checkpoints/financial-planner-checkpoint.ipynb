### Collect Crypto Prices Using the `requests` Library

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=USD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=USD"
```

```python
# Fetch current BTC price
current_BTC_price = requests.get(btc_url).json()
current_BTC_price

# Fetch current ETH price
current_ETH_price = requests.get(eth_url).json()
current_ETH_price

# Compute current value of my crpto
current_BTC_value = current_BTC_price['data']['1']['quotes']['USD']['price']
current_ETH_value = current_ETH_price['data']['1027']['quotes']['USD']['price']

# Print current crypto wallet balance
print(f"The current value of your {BTC} BTC is ${current_BTC_value:0.2f}")
print(f"The current value of your {ETH} ETH is ${current_ETH_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
APCA_API_KEY_ID = os.getenv("APCA_API_KEY")
APCA_API_SECRET_KEY = os.getenv("APCA_SECRET_KEY")
ALPACA_ENDPOINT_KEY = os.getenv("ALPACA_END_POINT")

#print(os.getenv("APCA_API_KEY_ID"))
#print(os.getenv("APCA_API_SECRET_KEY"))
#print(os.getenv("ALPACA_ENDPOINT_KEY"))

# Create the Alpaca API object
alpaca = tradeapi.REST(APCA_API_KEY_ID, APCA_API_SECRET_KEY, api_version="v2",base_url= ALPACA_ENDPOINT_KEY)
```

```python
# Format current date as ISO format
current_Date = pd.Timestamp("2023-10-05", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
historical_Data_df = alpaca.get_bars(tickers, timeframe, start = current_Date, end = current_Date, limit = 1000).df

# Reset the index
historical_Data_df.reset_index(inplace=True)

# Pivot the DataFrame
historical_Data_df = historical_Data_df.pivot(index='timestamp', columns='symbol')

# Reset the column names
historical_Data_df.columns = historical_Data_df.columns.swaplevel(0, 1)

# Reorder the columns as desired
column_order = ['AGG', 'SPY']
historical_Data_df = historical_Data_df[column_order]

# Drop the 'vwap' and 'trade_count' columns literally don't know why I had those
historical_Data_df = historical_Data_df.drop(columns=[('AGG', 'vwap'), ('SPY', 'vwap'), ('AGG', 'trade_count'), ('SPY', 'trade_count')])

# Display the updated DataFrame
historical_Data_df

```

```python
# Pick AGG and SPY close prices
# Extract and convert the closing prices
agg_close_price = float(historical_Data_df['AGG']['close'])
spy_close_price = float(historical_Data_df['SPY']['close'])

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_spy_value = my_spy * spy_close_price
my_agg_value = my_agg * agg_close_price

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income = 12000

# Consolidate financial assets data
cryptoAssets = current_BTC_value + current_ETH_value  #adding together all crypto
stock_bondAssets = my_agg_value + my_agg_value #adding together all stocks and bonds

# Create savings DataFrame
financial_Data = pd.DataFrame({'asset': ["crypto", "shares"],'amount': [cryptoAssets, stock_bondAssets]})

# Create savings DataFrame
df_savings = pd.DataFrame(financial_Data).set_index('asset')

# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
# Define custom colors
custom_colors = ['#ff9999', '#66b3ff']

# Create a pie chart with custom colors
pie_chart = df_savings.plot.pie(y='amount', title="Savings Portfolio Composition", figsize=(5, 5), colors=custom_colors)
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
total_Savings = df_savings['amount'].sum()

# Validate saving health
# Check if total savings are greater than or equal to the emergency fund
if total_Savings >= emergency_fund:
    if total_Savings > emergency_fund:
        print(f"Congratulations! You have enough money in your emergency fund.")
    else:
        print(f"You have reached your financial goal!")
else:
    print(f"You're ${emergency_fund - total_Savings:0.2f} away from achieving your financial goal. Keep saving!")
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date1 = pd.Timestamp('2018-10-08', tz='America/New_York').isoformat()
end_date1 = pd.Timestamp('2023-10-08', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG

df_portfolio_complete = alpaca.get_bars(tickers,timeframe,start=start_date1,end=end_date1,limit=3000).df

# Reset the index
df_portfolio_complete.reset_index(inplace=True)

# Pivot the DataFrame
df_portfolio_complete = df_portfolio_complete.pivot(index='timestamp', columns='symbol')

# Reset the column names
df_portfolio_complete.columns = df_portfolio_complete.columns.swaplevel(0, 1)

# Drop the 'vwap' and 'trade_count' columns literally don't know why I had those
df_portfolio_complete = df_portfolio_complete.drop(columns=[('AGG', 'vwap'), ('SPY', 'vwap'), ('AGG', 'trade_count'), ('SPY', 'trade_count')])

# Reorder the columns as desired
column_order = ['AGG', 'SPY']
df_portfolio_complete = df_portfolio_complete[column_order]

# Display sample data
df_portfolio_complete.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# Set number of simulations
num_sims = 500

num_trading_days = 252 * 30

MC_simulator = MCSimulation(portfolio_data = df_portfolio_complete, num_simulation = num_sims, num_trading_days = num_trading_days)

pd.set_option('display.float_format', '{:.6f}'.format)

MC_simulator.portfolio_data
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_simulator.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_simulator.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_simulator.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
summary_Table = MC_simulator.summarize_cumulative_return()

# Print summary statistics
print(summary_Table)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(summary_Table[8]*initial_investment,2)
ci_upper = round(summary_Table[9]*initial_investment,2)


# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower = round(summary_Table[8]*initial_investment,2)
ci_upper = round(summary_Table[9]*initial_investment,2)


# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
